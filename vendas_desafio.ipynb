# Extração dos Dados

```python
import pandas as pd

nome_arquivo = 'C:/Users/Liliane/Documents/Cursos Liliane/Bootcamp Santander/Desafio1/para entregar/vendas2.csv'

# Nome da coluna que contém as datas
nome_coluna_data = 'DATAM'
nome_coluna_valor = 'VALOR'


df = pd.read_csv(nome_arquivo, delimiter=';', encoding='latin1')

# Acesse as colunas de data e valores de venda
datas = df[nome_coluna_data]
valores = df[nome_coluna_valor]

# Exibindo para ver se ficou ok
print(datas, valores)


```

# Transformação

```python
# Criar um DataFrame com as datas e valores
df_resultado = pd.DataFrame({'Data': datas, 'Valor': valores})

# Agrupe os valores por data e calcular a soma
resultado = df_resultado.groupby('Data')['Valor'].agg('sum').reset_index()

# O resultado é um DataFrame com as datas e as somas correspondentes
print(resultado)
```

# Carregando o resultado e mostrando em grafico

```python
import matplotlib.pyplot as plt


df_grafico = pd.DataFrame(resultado)


# Criar um gráfico de barras
plt.figure(figsize=(10, 6))  # Defina o tamanho da figura (opcional)

plt.bar(df_grafico['Data'], df_grafico['Valor'])

# Configurar rótulos e título
plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Gráfico de Barras dos Valores por Data')

# Rotacionar os rótulos do eixo x para melhor legibilidade (opcional)
plt.xticks(rotation=0)

# Exibir o gráfico
plt.show()
```
